# Global Digital Artist Ranking

```python
# Library
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Sending Request
page_url = "https://kworb.net/itunes/extended.html"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0'
}

page = requests.get(page_url,
                    headers = headers)
print(page)

# Beautiful Soup Object
soup = BeautifulSoup(page.content,'html.parser')
soup

# Our Table Artist Ranking
table_rank = soup.find('table',
                   {'class':'sortable'})
table_rank

# Table Headers Artist Ranking
headers = []
for i in table_rank.find_all('th'):
    col_name = i.text
    headers.append(col_name)
    
# Data Frame Columns Name
rank = pd.DataFrame(columns=headers)
rank


# Table Data Artist Ranking
for i in table_rank.find_all('tr')[1:]:
    row_data = i.find_all('td')
    rows = [t.text for t in row_data]
    lenght = len(rank)
    rank.loc[lenght] = rows
rank
```

```python
# Storing into the excel file 
dfRanking.to_excel("Global Digital Artist Ranking & Diffrent Platforms.xlsx",
            index=False)

rank.to_excel("Global Digital Artist Ranking.xlsx",
                      index=False)
```
