# Spotify most streamed artists of all time

```python
# Library
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Sending Request
page_url = "https://kworb.net/spotify/artists.html"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0'
}

page = requests.get(page_url,
                    headers = headers)
print(page)

# Beautiful Soup Object
soup = BeautifulSoup(page.content,'html.parser')
soup

# Our Table Spotify most streamed artists of all time
table_all_time = soup.find('table',
                   {'class':'addpos sortable'})
table_all_time

# Table Headers Spotify most streamed artists of all time
headers = []
for i in table_all_time.find_all('th'):
    col_name = i.text
    headers.append(col_name)
    
# Data Frame Columns Name
dftable_all_time = pd.DataFrame(columns=headers)
dftable_all_time

# Table Data Spotify most streamed artists of all time
for i in table_all_time.find_all('tr')[1:]:
    row_data = i.find_all('td')
    rows = [t.text for t in row_data]
    lenght = len(dftable_all_time)
    dftable_all_time.loc[lenght] = rows
dftable_all_time
```

# Spotify most streamed songs

```python
# Library
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Sending Request
page_url = "https://kworb.net/spotify/toplists.html"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0'
}

page = requests.get(page_url,
                    headers = headers)
print(page)

# Beautiful Soup Object
soup = BeautifulSoup(page.content,'html.parser')
soup

# Our Table Spotify most streamed songs
table_streamed = soup.find('table',
                   {'style':'width: 650px;'})

table_streamed


# Table Headers Spotify most streamed songs
headers = []
for i in table_streamed.find_all('th'):
    col_name = i.text
    headers.append(col_name)

# Data Frame Columns Name
dftable_streamed = pd.DataFrame(columns=headers)
dftable_streamed


# Table Data Spotify most streamed songs
for i in table_streamed.find_all('tr')[1:]:
    row_data = i.find_all('td')
    rows = [t.text for t in row_data]
    lenght = len(dftable_streamed)
    dftable_streamed.loc[lenght] = rows
    
dftable_streamed
```
